```python
ReviewId=[]
AssessmentId=[]
Workload=[]
Difficulty=[]
Rating=[]
TextReviews=[]
for reviewId in reviewid_crn:
    assessments_id=crn_cs_assess[crn_cs_assess['CRN']==reviewId[1]]['AssessmentId'].tolist()
    for i in range(len(assessments_id)):
        ReviewId.append(reviewId[0])
    for ass_id in assessments_id:
        AssessmentId.append(ass_id)
        Workload.append(random.randint(1,5))
        Difficulty.append(random.randint(1,5))
        r=random.choice(text_review)
        TextReviews.append(r[0])
        Rating.append(r[1])
    
```

```python
sub_review_dict={'ReviewId':ReviewId,'AssessmentId':AssessmentId,'Workload':Workload,'Difficulty':Difficulty,'Rating':Rating,'Text Reviews':TextReviews}
```

```python
Sub_Reviews=pd.DataFrame(sub_review_dict)
```
