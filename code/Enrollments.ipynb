```python
grades_list=['A+','A','A-','B+','B','B-','C+','C','C-','D+','D','D-','F']
```

```python
Term,CRN,NetId,Credits,Grade,Score  = [[] for k in range(0,6)] 
for i in range(len(ct)):
    Term1 = ct[i][1]
    Term.append(Term1)
    CRN1=ct[i][0]
    CRN.append(CRN1)
    rand = random.randint(3,4)
    Credits1 = str(rand)
    Credits.append(Credits1)
    Grade1=str(random.choice(grades_list))
    Grade.append(Grade1)
    
d = {"Term":Term,"CRN":CRN ,"Credits":Credits,'Grade':Grade}
df = pd.DataFrame(d)
df
    
```

```python
df['Score']=''
```
